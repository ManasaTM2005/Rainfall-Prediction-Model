```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LinearRegression

# ---- Sample data ----
sample_df = df_select.sample(n=min(10000, len(df_select)), random_state=42)

# Convert datetime column into numeric features
sample_df['dt_txt'] = pd.to_datetime(sample_df['dt_txt'])
sample_df['year'] = sample_df['dt_txt'].dt.year
sample_df['month'] = sample_df['dt_txt'].dt.month
sample_df['day'] = sample_df['dt_txt'].dt.day
sample_df['hour_from_dt'] = sample_df['dt_txt'].dt.hour

# X and y
X_sample = sample_df.drop(['label_rain_24h', 'dt_txt'], axis=1)
y_sample = sample_df['label_rain_24h']

# Encode city
le = LabelEncoder()
X_sample['city'] = le.fit_transform(X_sample['city'].astype(str))

# Scale
scaler = StandardScaler()
X_sample_scaled = scaler.fit_transform(X_sample)

# ---- Train model ----
model_lr = LinearRegression()
model_lr.fit(X_sample_scaled, y_sample)

# ---- Feature importance plot ----
importances = np.abs(model_lr.coef_)
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(12, 6))
plt.title("Feature Importance (Your Dataset - Linear Regression)")

plt.bar(range(len(X_sample.columns)), importances[indices], align="center")
plt.xticks(range(len(X_sample.columns)), X_sample.columns[indices], rotation=90)

plt.ylabel("Absolute Coefficient Magnitude")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# ---- Optimize data ----
q_low = df_select['wind_speed'].quantile(0.0)
q_high = df_select['wind_speed'].quantile(0.85)

df_super_opt = df_select[(df_select['wind_speed'] >= q_low) &
                         (df_select['wind_speed'] <= q_high)]

print(f"Original Data Size: {df_select.shape}")
print(f"Super Optimized Data Size (85% retention): {df_super_opt.shape}")
print(f"Max Rainfall in optimized data: {df_super_opt['wind_speed'].max()}")

# ---- Convert dt_txt to numeric ----
df_super_opt['dt_txt'] = pd.to_datetime(df_super_opt['dt_txt'])

df_super_opt['year'] = df_super_opt['dt_txt'].dt.year
df_super_opt['month'] = df_super_opt['dt_txt'].dt.month
df_super_opt['day'] = df_super_opt['dt_txt'].dt.day
df_super_opt['hour_from_dt'] = df_super_opt['dt_txt'].dt.hour

# ---- X and y ----
X_super = df_super_opt.drop(['wind_speed', 'dt_txt'], axis=1)
y_super = df_super_opt['wind_speed']

# Encode city if it is text
if 'city' in X_super.columns:
    le = LabelEncoder()
    X_super['city'] = le.fit_transform(X_super['city'].astype(str))

# ---- Train-test split ----
X_train, X_test, y_train, y_test = train_test_split(
    X_super, y_super, test_size=0.2, random_state=42
)

# ---- Model ----
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# ---- Prediction ----
y_pred = lr_model.predict(X_test)

# ---- Metrics ----
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("\nRegression Evaluation Metrics:")
print(f"MAE: {mae:.4f}")
print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R2 Score: {r2:.4f}")

# ---- Graph ----
plt.figure(figsize=(6, 4))
bars = plt.bar(['MSE', 'RMSE', 'MAE'], [mse, rmse, mae])
plt.title('Evaluation Metrics (Super Optimized)')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2),
             va='bottom', ha='center')

plt.show()
```

```python
def predict_rainfall(input_data):
    
    input_df = pd.DataFrame([input_data])
    
    try:
        prediction = lr_model.predict(input_df)
        return prediction[0]
    except Exception as e:
        return f"Error: {str(e)}"


print("--- Prediction Pipeline Test ---")
sample_input = X_test.iloc[0].to_dict()
print("Input Features:", sample_input)
predicted_val = predict_rainfall(sample_input)
print(f"\nPredicted Rainfall: {predicted_val:.2f} mm")
print(f"Actual Rainfall: {y_test.iloc[0]:.2f} mm")
```
